This script is just used to combine a group of model IDs into a list. This is meant to group all models with the same architecture and training hyperparameters that differ just in the random seed used for parameter initialization.

This is necessary because all models are meant to be saved in `\results\models\`, but during the baseline exploration I want to analyse specific subsets of models that have something in common.

In a future iteration the models might be saved in `\results\models\{experiment}\` with `{experiment}` being a directory that already groups a subset of models for some reason.

```python
import os
import numpy as np
```

```python
project_root = os.path.abspath(os.path.join(os.getcwd(), '../'))
```

```python
target_dir = 'models_baseline_1'
```

I use `target_dir` as a temporary directory inside `\results\models\` where I put all the models that should be grouped together. A list with their IDs will then be saved in `\data\processed`. `target_dir` is mean to be deleted after this, with its models being returned to `\results\models\`.

```python

models = os.listdir(os.path.join(project_root, 'results', target_dir))
models
```

```python
file_name = 'models_IDs_baseline_1.npy'
```

```python
with open(os.path.join(project_root, 'data', 'processed', file_name), 'wb') as f:
    np.save(f, np.array(models))
```
