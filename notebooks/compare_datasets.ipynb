```python
import torch, os, sys

project_root = os.path.abspath(os.path.join(os.getcwd(), '../'))
sys.path.append(project_root)
from src.data import LanderDataset
```

```python
seq_len = range(4, 22, 2)
dataset_size = []
for sl in seq_len:
    train_dataset = torch.load(os.path.join(project_root, 'data', 'processed', f'gymnasium-ActorCritic-LunarLander-1000-{sl}-os.pt'), weights_only=False)
    print(sl, len(train_dataset))
    dataset_size.append(len(train_dataset))
```

```python
import numpy as np
for i in range(1, len(dataset_size)):
    diff = np.round(((dataset_size[i]-dataset_size[i-1])/(dataset_size[i-1]))*100, 2)
    print(f'{seq_len[i-1]} -> {seq_len[i]} = {diff}%')
```

```python
import pandas as pd

df = pd.DataFrame({'sequence length': seq_len, 'dataset size': dataset_size})
ax = df.plot.bar(x='sequence length', y='dataset size')
```
